#Solve this problem using Logistic Regression(using numpy from scratch)

```python
def sigmoid(z):
  return 1/(1+np.exp(-z))
```

```python
class logistic_regression:
  def __init__(self,lr=0.01, epochs=1000, reg_strength=0.0 ):
    self.lr=lr
    self.epochs = epochs
    self.reg_strength = reg_strength
    self.losses=[]
  def fit(self, X, y):
        m, n = X.shape
        self.theta = np.zeros(n)

        for epoch in range(self.epochs):
            z = np.dot(X, self.theta)
            h = sigmoid(z)
            # Loss (with L2 regularization)
            loss = (-1/m) * (np.dot(y, np.log(h+1e-9)) + np.dot((1-y), np.log(1-h+1e-9))) \
                   + (self.reg_strength/(2*m))*np.sum(self.theta**2)
            self.losses.append(loss)

            # Gradient
            grad = (1/m) * np.dot(X.T, (h-y)) + (self.reg_strength/m)*self.theta

            # Update rule
            self.theta -= self.lr * grad
        return self.losses

  def predict(self, X):
        return sigmoid(np.dot(X, self.theta)) >= 0.5


```

```python
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix
from sklearn.preprocessing import StandardScaler
# Scaling
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
model = logistic_regression(lr=0.01, epochs=1000, reg_strength=0.0)
model.fit(X_train, y_train)

y_pred = model.predict(X_test)

print("Accuracy:", accuracy_score(y_test, y_pred))
print("Precision:", precision_score(y_test, y_pred))
print("Recall:", recall_score(y_test, y_pred))
print("F1 Score:", f1_score(y_test, y_pred))
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred))
```

```python
for lam in [0.0, 0.1, 1.0]:
    model = logistic_regression(lr=0.1, epochs=1000, reg_strength=lam)
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    print(f"\nλ={lam}")
    print("Accuracy:", accuracy_score(y_test, y_pred))
    print("Precision:", precision_score(y_test, y_pred))
    print("Recall:", recall_score(y_test, y_pred))
    print("F1 Score:", f1_score(y_test, y_pred))
```

```python
# Train for λ=1.0
model = logistic_regression(lr=0.1, epochs=1000, reg_strength=1.0)
losses = model.fit(X_train, y_train)

# Plot loss curve
plt.figure(figsize=(8,5))
plt.plot(range(len(losses)), losses, label="λ=1.0")
plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.title("Learning Dynamics: Loss Curve (λ=1.0)")
plt.legend()
plt.grid(True)
plt.show()

```

```python
lambdas = [0.0, 0.1, 1.0, 10.0, 100.0, 500.0]
accuracies, precisions, recalls, f1s = [], [], [], []

for lam in lambdas:
    model = logistic_regression(lr=0.1, epochs=1000, reg_strength=lam)
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)

    accuracies.append(accuracy_score(y_test, y_pred))
    precisions.append(precision_score(y_test, y_pred))
    recalls.append(recall_score(y_test, y_pred))
    f1s.append(f1_score(y_test, y_pred))

# Plot all metrics
plt.figure(figsize=(8,5))
plt.plot(lambdas, accuracies, marker='o', label="Accuracy")
plt.plot(lambdas, precisions, marker='o', label="Precision")
plt.plot(lambdas, recalls, marker='o', label="Recall")
plt.plot(lambdas, f1s, marker='o', label="F1 Score")

plt.xscale("log")  # log scale for lambda
plt.xlabel("λ (Regularization Strength)")
plt.ylabel("Score")
plt.title("Effect of Regularization on Logistic Regression (Spam Dataset)")
plt.legend()
plt.grid(True)
plt.show()
```
